# MIDIA / DIA-PASEF interactive clustering

Load a retention-time slice of a timsTOF run, then tune HDBSCAN parameters live to cluster the precursor (MS1, 3D) or MIDIA fragment (4D) point clouds. Backed by `imspy-core`; served with Voila.

Set `RUN_PATH` below to a `.d` folder, then use the **Data** tab to load a slice and the cluster tabs to explore.

```python
import warnings
warnings.filterwarnings('ignore')

import os
from imspy_vis.midia import MidiaVis

RUN_PATH = os.environ.get(
    'MIDIA_RUN_PATH',
    '/Users/davidteschner/Promotion/midia/timsim/DIA-250K-NOISY-CHRONO/DIA-250K-NOISY-CHRONO.d',
)
# Folder the run browser opens in (a directory of runs); override with MIDIA_BROWSE_DIR.
BROWSE_DIR = os.environ.get('MIDIA_BROWSE_DIR', '/Users/davidteschner/Promotion/midia/timsim/')

app = MidiaVis(default_path=RUN_PATH, browse_dir=BROWSE_DIR)
app.display()
```
